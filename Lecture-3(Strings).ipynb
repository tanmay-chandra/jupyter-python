```python
#Indexing
#Positive indexing
word = "Python"
print(len(word))
print(word[0])
print(word[1])
print(word[5])
```

```python
#Negative indexing
word = "Python"
print(len(word))
print(word[-1])
print(word[-2])
print(word[-4])
print(word[-6])
```

```python
# Slicing
word = "python"
#print(word[start:stop:step])
#start = 0
#stop = len(word)
#step = 1
print(word[::])
print(word[0:6:])
print(word[1:6:1])
print(word[0:2:1])
print(word[0:6:2])
print(word[4::])
```

```python
#Negative slicing
word = "python"
print(word[-2::])
print(word[-6::1])
print(word[-6:-4:])
```

```python
# String formatting
name = "Sachin"
city = "noida"
age = 23
print("hi name your city is city and age is age.")

print(f"\nhi {name} your city is {city} and age is {age}")

print("\nhi {} your city is {} and age is {}".format(name,city,age))

print("\nhi {1} your city is {2} and age is {0}".format(age,name,city))

print("hi %s your city is %s and age is %s"%(name,city,age))
```
