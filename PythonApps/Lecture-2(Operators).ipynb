# Identity operators

```python
a=5
b=500
c=58
```

```python
print(a is b)
```

```python
print(id(a))
print(id(b))
print(id(c))
```

```python
a=[1,100,200]
b=[1,100,200]

print(a==b)     # checks values
print(a is b)   # checks memory location

print(id(a))
print(id(b))
```

```python
a=5
b=5
print(a is not b)
```
