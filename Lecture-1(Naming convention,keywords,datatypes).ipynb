#                                               RULES
# 1. variable names must start with a letter (a-z,A-Z) or an underscore(_).
# 2. python is case-sensitive.so,name and Name would be considered different variables.
# 3. variables names cannot be the same as Python keywords or reserved words.
# examples include if,else,for,while,True,False,etc.
# 4. for multi-word variable names, use underscores to seperate words (snake_case convention). for example , my_variable,user_name
